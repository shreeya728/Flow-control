```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

file_path = ['btp_analysis_1.xlsx', 'btp_analysis_2.xlsx', 'btp_analysis_3.xlsx', 'btp_analysis_4.xlsx', 'btp_analysis_5.xlsx', 'btp_analysis_6.xlsx', 'btp_analysis_7.xlsx', 'btp_analysis_8.xlsx', 'btp_analysis_9.xlsx']
sheet_name = ['r1_2k', 'r1_2.5k', 'r1_3k', 'r1_3.5k', 'r1_4k']
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'purple']  # Define colors for each set
labels = ['Set 1', 'Set 2', 'Set 3', 'Set 4', 'Set 5', 'Set 6', 'Set 7', 'Set 8', 'Set 9']  # Define labels for each set

# Initialize data lists outside the loops
all_data = []

for j in range(len(sheet_name)):
    data_for_sheet = []  # Data for the current sheet
    for i in range(len(file_path)):
        if os.path.exists(file_path[i]):  # Check if file exists
            try:
                df = pd.read_excel(file_path[i], sheet_name=sheet_name[j])
                fifth_column = df.iloc[:, 2]
                filtered_column = fifth_column[fifth_column != 0]
                mean = filtered_column.mean()
                data_for_sheet.append(mean)  # Append mean to the current sheet's data
            except ValueError:
                print(f"Sheet '{sheet_name[j]}' not found in '{file_path[i]}'")
    all_data.append(data_for_sheet)  # Append the sheet's data to the main data list

# Plotting
x = [2000, 2500, 3000, 3500, 4000]

for i, data_set in enumerate(all_data):
    x_values = [x[i]] * len(data_set)  # Repeat x-value for each data set
    for k in range(len(data_set)):
        plt.scatter(x_values[k], data_set[k], color=colors[k], label=labels[k] if i == 0 else None)  # Plot with color and label


plt.ylabel("Pitot tube")
plt.xlabel("Fan RPM")
plt.title("Velocity at different fan speeds")
plt.grid(True)
plt.legend()  # Show legend
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'btp_analysis_6.xlsx'
sheet_name = 'r2_4k'  # Replace with the actual sheet name

df = pd.read_excel(file_path, sheet_name=sheet_name)

time = df.iloc[:, 0]  # Assuming the first column is time
sec_column = df.iloc[:, 2]

# Remove the first value from time and fifth column
time = time[1:]
sec_column = sec_column[1:]

# Convert 'time' column to numeric, handling errors
time = pd.to_numeric(time, errors='coerce')

# Calculate incremental steps starting from 0, ignoring NaNs
time = time - time.iloc[0]

# Create the plot
plt.plot(time, sec_column)

# Set labels and title
plt.xlabel("Number of samples")
plt.ylabel("Velocity")
plt.title("Velocity vs No. of samples")

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```
